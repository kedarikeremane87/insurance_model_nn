```python
x = df.drop('charges', axis = 'columns')
y = df['charges']
```

```python
#encoders

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
gen_encoder = LabelEncoder()
smoker_encoder = LabelEncoder()
region_encoder = OneHotEncoder()
```

```python
df['sex'] = gen_encoder.fit_transform(df['sex'])
df['smoker'] = smoker_encoder.fit_transform(df['smoker'])
df.head()
```

```python
region_out = region_encoder.fit_transform(df[['region']]).toarray()
region_out
```

```python
region_out = pd.DataFrame(region_out, columns=region_encoder.get_feature_names_out())
region_out.head()
```

```python
encoded_df = pd.concat([df, region_out], axis=1)
encoded_df.head()
```

```python
encoded_df = encoded_df.drop('region', axis = 1)
encoded_df.head()
```

```python
x = encoded_df.drop('charges', axis =1)
y=encoded_df['charges']
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.75)
```

```python
print(x.shape)
print(y.shape)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
%pip install tensorflow
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(100,activation='relu',input_shape=(9,)))
model.add(Dense(100,activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(xtrain,ytrain,epochs=5000)
```

```python
model.save("./Day 3 Insurance NN/insurance.h5")
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
print(np.sqrt(mean_squared_error(ytest, ypred)))
```

```python
print(r2_score(ytest, ypred))
```
